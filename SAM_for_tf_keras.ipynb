```python
def basic_augmentation(x):
    x = tf.image.random_flip_left_right(x)
    x = tf.pad(x, [ [4,4], [4,4], [0,0]], "reflect")
    x = tf.image.random_crop(x, [32,32,3])
    return x


def cutout_augmentation(image, cval=0, cnum = 6, csize = 0.25):
    DIM = image.shape[0]
    for k in range( cnum ):
        # CHOOSE RANDOM LOCATION
        x = tf.cast( tf.random.uniform([],0,DIM),tf.int32)
        y = tf.cast( tf.random.uniform([],0,DIM),tf.int32)
        # COMPUTE SQUARE 
        WIDTH = tf.cast( csize*DIM,tf.int32)
        ya = tf.math.maximum(0,y-WIDTH//2)
        yb = tf.math.minimum(DIM,y+WIDTH//2)
        xa = tf.math.maximum(0,x-WIDTH//2)
        xb = tf.math.minimum(DIM,x+WIDTH//2)
        # DROPOUT IMAGE
        one = image[ya:yb,0:xa,:]
        two = tf.fill([yb-ya,xb-xa,3], tf.cast(cval, image.dtype) ) 
        three = image[ya:yb,xb:DIM,:]
        middle = tf.concat([one,two,three],axis=1)
        image = tf.concat([image[0:ya,:,:],middle,image[yb:DIM,:,:]],axis=0)

    # RESHAPE HACK SO TPU COMPILER KNOWS SHAPE OF OUTPUT TENSOR 
    image = tf.reshape(image,[DIM,DIM,3])
    return image


def make_dataset( train_data, validation_data, batch_size,cutout_num=1,cutout_size=0.25):
    (x_train, label_train)= train_data
    (x_test, label_test)=validation_data

    train_len = len(x_train)
    test_len = len(x_test)

    ds_train = tf.data.Dataset.from_tensor_slices(train_data)
    ds_train = ds_train.shuffle(train_len).batch(batch_size, drop_remainder=True)
    ds_validation = tf.data.Dataset.from_tensor_slices(validation_data)
    ds_validation = ds_validation.batch(batch_size)

    def data_augmentation(images,labels,aug=True,cutout_num=1,cutout_size=0.25):
        image_mean = tf.constant([[[0.49139968, 0.48215841, 0.44653091]]])
        image_std = tf.constant([[[0.24703223, 0.24348513, 0.26158784]]])
        images = tf.cast(images, tf.float32)/255.0
        images = (images-image_mean)/image_std
        if aug:
            images = tf.map_fn(basic_augmentation, images)
            if cutout_num!=0:
                images = tf.map_fn(lambda image: cutout_augmentation(image,cval=0.0, csize=cutout_size, cnum=cutout_num), images)

        return images, labels

    ds_train = ds_train.map(lambda image, label: data_augmentation(image, label, True, cutout_num,cutout_size  ), num_parallel_calls=tf.data.experimental.AUTOTUNE)
    ds_train = ds_train.prefetch(tf.data.experimental.AUTOTUNE)

    ds_validation = ds_validation.map(lambda image, label: data_augmentation(image, label, False ), num_parallel_calls=tf.data.experimental.AUTOTUNE)
    ds_validation = ds_validation.prefetch(tf.data.experimental.AUTOTUNE)
    return ds_train,ds_validation

class LRSchedule(tf.keras.optimizers.schedules.LearningRateSchedule):
    def __init__(self, **kwargs):
        super().__init__()
        self.lr = kwargs['lr']
        self.total_steps = kwargs['total_steps']
        self.warmup = kwargs['warmup']
        self.cooldown = kwargs['cooldown']

    def get_config(self):
        config = {"lr": float(self.lr), 
                  "warmup": float(self.warmup),
                  "cooldown": float(self.cooldown),
                  "total_steps": float(self.total_steps),
                  }
        return config

    def __call__(self, step):
        step_l = step-self.warmup
        flat_steps = self.total_steps-self.warmup-self.cooldown
        total_l = self.total_steps-self.warmup
        lr2 = tf.where(step <= self.warmup, step*self.lr/self.warmup, self.lr)
        return tf.where(step_l <= total_l-self.cooldown,
                        lr2,
                        self.lr*(1/2*(1+tf.math.cos((step-(self.total_steps-self.cooldown))*np.pi/self.cooldown)))
            )

def showHistory(history):
    import matplotlib.pyplot as plt

    # Setting Parameters
    acc = history['acc']
    val_acc = history['val_acc']
    max_acc = max(val_acc)

    loss = history['loss']
    val_loss = history['val_loss']

    epochs = range(len(acc))

    plt.figure(figsize=(16,6))
 
    # Accracy
    plt.subplot(1,2,1)
    plt.plot(epochs, acc, 'r', label='Training')
    plt.plot(epochs, val_acc, 'b', label='Validation')
    plt.title('Accuracy')
    plt.grid()
    plt.legend()

    # Loss 
    plt.subplot(1,2,2)
    plt.plot(epochs, loss, 'r', label='Training')
    plt.plot(epochs, val_loss, 'b', label='Validation')
    plt.title('Loss')
    plt.ylim(0.0,2.0)
    plt.grid()
    plt.legend()
    plt.show()
```

```python
import pickle
from tensorflow.keras import layers
import tensorflow_addons as tfa

# https://github.com/google-research/sam
class SAMModel(tf.keras.models.Model):
    def __init__(self, *args, rho=0.05, **kwargs):
        super().__init__( *args, **kwargs)
        self.rho = tf.constant(rho, dtype=tf.float32)

    def train_step(self,data):
        x, y = data

        # 1st step
        with tf.GradientTape() as tape:
            y_pred = self(x, training=True)
            loss = self.compiled_loss(y, y_pred, regularization_losses=self.losses)
        
        trainable_vars = self.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)
        
        norm = tf.linalg.global_norm(gradients)
        scale = self.rho / (norm + 1e-12)
        e_w_list = []
        for v, grad in zip(trainable_vars, gradients):
            e_w = grad * scale
            v.assign_add(e_w)
            e_w_list.append(e_w)

        # 2nd step
        with tf.GradientTape() as tape:
            y_pred_adv = self(x, training=True)
            loss_adv = self.compiled_loss(y, y_pred_adv, regularization_losses=self.losses)
        gradients_adv = tape.gradient(loss_adv, trainable_vars)
        for v, e_w in zip(trainable_vars, e_w_list):
            v.assign_sub(e_w)

        # optimize
        self.optimizer.apply_gradients(zip(gradients_adv, trainable_vars))

        self.compiled_metrics.update_state(y, y_pred)
        return {m.name: m.result() for m in self.metrics}

def SAM_wide_res_net(depth,width_factor, input_shape, num_outputs, l2_reg=0.0, rho=0.0):
    filters = [16, 1 * 16 * width_factor, 2 * 16 * width_factor, 4 * 16 * width_factor]
    n = int((depth-4)/6)

    CONV_KWARGS = {
        'use_bias': False,
        'padding': 'same',
        'kernel_initializer': tf.keras.initializers.VarianceScaling(scale=2.0, mode='fan_out', distribution='normal'),
        'kernel_regularizer': tf.keras.regularizers.l2(l2_reg)
    }
    BN_KWARGS = {
        'momentum': 0.9,
        'epsilon' : 1e-5         
    }
    init_range = 1.0 / (num_outputs)**(0.5)
    DENSE_KWARGS = {
        'kernel_initializer' : tf.keras.initializers.RandomUniform(minval=-init_range, maxval=init_range)    ,     
        'kernel_regularizer': tf.keras.regularizers.l2(l2_reg),
        'bias_regularizer': tf.keras.regularizers.l2(l2_reg)
    }
    def res_basic(x, filters, strides, pre_act=False):
        shortcut = x
        if pre_act:
            x = layers.BatchNormalization( **BN_KWARGS )(x)
            x = layers.ReLU()(x)
        x = layers.Conv2D(filters, 3, strides,**CONV_KWARGS)(x)
        x = layers.BatchNormalization( **BN_KWARGS )(x)
        x = layers.ReLU()(x)
        x = layers.Conv2D(filters, 3, 1,**CONV_KWARGS)(x)
        x = layers.BatchNormalization( **BN_KWARGS )(x)
        x = layers.ReLU()(x)
        
        shortcut_shape = K.int_shape(shortcut)
        x_shape = K.int_shape(x)
        downsample_strides = 2 if shortcut_shape[1] != x_shape[1] else 1
        if downsample_strides != 1:
            shortcut = layers.AveragePooling2D((downsample_strides,downsample_strides), downsample_strides)(shortcut)
        if shortcut_shape[3] != x_shape[3]:
            shortcut = tf.pad(shortcut, [ [0,0],[0,0],[0,0],[0,x_shape[3]-shortcut_shape[3]] ] )

        return shortcut+x

    x = inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(filters[0], 3, 1, **CONV_KWARGS)(x)
    for i in range(n):
        x = res_basic(x, filters[1], strides=1, pre_act = (i == 0) )
    for i in range(n):
        x = res_basic(x, filters[2], strides=2 if i == 0 else 1 )
    for i in range(n):
        x = res_basic(x, filters[3], strides=2 if i == 0 else 1 )
    x = layers.BatchNormalization( **BN_KWARGS )(x)
    x = layers.ReLU()(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(units=num_outputs, **DENSE_KWARGS)(x)
    x = layers.Activation('softmax')(x)
    if rho==0.0:
        model = tf.keras.models.Model(inputs,x) 
    else:
        model = SAMModel(inputs,x, rho=rho) 
    model.build(input_shape)
    return model

def train(model, optimizer, callbacks, epochs, ds_train, ds_validation,verbose=1, label_smoothing=0.0):
    
    loss = tf.keras.losses.CategoricalCrossentropy(label_smoothing=label_smoothing)
    acc = tf.keras.metrics.CategoricalAccuracy(name='acc')
    model.compile(optimizer=optimizer,
                loss=loss,metrics=[acc], steps_per_execution=10)
    history = model.fit(ds_train,epochs=epochs,
                        validation_data=ds_validation,verbose=verbose, callbacks=callbacks)
    
    return history.history

def train_sam():
    wrn_depth = 22
    wrn_width_factor = 8

    warmup_epochs = 0
    cooldown_epochs = 200
    EPOCHS = warmup_epochs+cooldown_epochs

    BATCH_SIZE = 128
    max_lr = 0.1
    min_lr = max_lr/100
    wd = 5e-4
    label_smoothing = 0.0
    sam_rho_list = [0.05]

    cutout_num = 0
    cutout_size = 0.5

    def scheduler(epoch, lr):
        flat_epochs = EPOCHS - warmup_epochs - cooldown_epochs
        if epoch < warmup_epochs:
            return min_lr + (max_lr-min_lr)*epoch/warmup_epochs
        elif epoch < warmup_epochs+flat_epochs:
            return max_lr
        else:
            epoch = epoch - (warmup_epochs+flat_epochs) + 1
            return min_lr + 0.5*(max_lr-min_lr)*(1.0+np.cos(epoch/cooldown_epochs*np.pi))
    class CustomLogger(tf.keras.callbacks.Callback):
        def on_epoch_end(self, epoch, logs):
            lr = K.get_value(self.model.optimizer.lr)
            logs['lr'] = lr

    for sam_rho in sam_rho_list:
        optimizer = tf.optimizers.SGD(max_lr, momentum=0.9, global_clipnorm=5.0)

        lr_scheduler = tf.keras.callbacks.LearningRateScheduler(scheduler)
        logger = CustomLogger()

        ds_train, ds_validation = make_dataset( train_data, validation_data, BATCH_SIZE,cutout_num,cutout_size)

        model = SAM_wide_res_net( wrn_depth, wrn_width_factor, (32,32,3), 10, wd, rho=sam_rho)

        start_time = time.time()
        start_datetime = datetime.datetime.now(tz=tz)
        print('Start:',start_datetime.isoformat())
        print('LR:',max_lr, 'WD:', wd ,'RHO:', sam_rho )
        history = train(  model,  optimizer, [lr_scheduler,logger], EPOCHS,
                        ds_train, ds_validation, 
                        verbose=1, label_smoothing=label_smoothing)
        end_datetime = datetime.datetime.now(tz=tz)
        print('End:',end_datetime.strftime("%Y/%m/%d %H:%M:%S"))
        print('Time:', time.time()-start_time )

        showHistory(history)
        f = open( f'{end_datetime.strftime("%Y%m%d%H%M%S")}.hist', 'wb')
        pickle.dump(history, f)

try:
    tpu = tf.distribute.cluster_resolver.TPUClusterResolver()  # TPU detection
    tf.config.experimental_connect_to_cluster(tpu)
    tf.tpu.experimental.initialize_tpu_system(tpu)
    strategy = tf.distribute .TPUStrategy(tpu)
except ValueError:
    strategy = tf.distribute.get_strategy()

with strategy.scope():
    train_sam()
```
